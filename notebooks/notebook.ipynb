```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets  import load_iris
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
X, y = load_iris(return_X_y=True, as_frame=True)

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=2126, test_size=0.2)

df_train = X_train.copy()
df_train['Specie'] = y_train
df_train
```

```python
#  Continuous values mean that Gaussian can be applied. Discrete require multinomial, as well as bernoulli

df_train.sample(10)
```

```python
df_train.describe().T
```

```python
df_train.hist()
plt.show()
```

```python
sns.pairplot(data=df_train, hue='Specie')
plt.show()
```

```python
X_train.corr()
```

```python
model = GaussianNB()
model.fit(X_train, y_train)
```

```python
y_hat = model.predict(X_test)

print(f"Accuracy: {accuracy_score(y_test, y_hat):.2f}")
```

```python
#  decision tree, random forest, naive bayes

MODELS = {
    'Decision Tree Classifier': DecisionTreeClassifier(),
    'RandomForestClassifier': RandomForestClassifier(),
    'Gaussian Naive Bayes': GaussianNB()
}
```

```python
y_hat_predict = dict()

for name, model in MODELS.items():
    model.fit(X_train, y_train)
    y_hat_predict[name] = model.predict(X_test)
```

```python
for name, y_hat_i in y_hat_predict.items():
    print(f"Classifier Name: {name} ---> Accuracy: {accuracy_score(y_test, y_hat_i):.2f}")
```
